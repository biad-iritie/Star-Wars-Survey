# Star Wars Survey

After the increase audience of *Star Wars*, the team at [FiveThirtyEight](https://fivethirtyeight.com) try to reply to this question : **does the rest of America realize that “The Empire Strikes Back” is clearly the best of the bunch?**.
So the team collect data addressing this question by survey *Star Wars Fans* using online tool SurveyMonkey. The result of tis collect is [here](https://github.com/fivethirtyeight/data/tree/master/star-wars-survey).

The data has several columns, including:

- RespondentID - An anonymized ID for the respondent (person taking the survey)
- Gender - The respondent's gender
- Age - The respondent's age
- Household Income - The respondent's income
- Education - The respondent's education level
- Location (Census Region) - The respondent's location
- Have you seen any of the 6 films in the Star Wars franchise? - Has a Yes or No response
- Do you consider yourself to be a fan of the Star Wars film franchise? - Has a Yes or No response

There are several other columns containing answers to questions about the Star Wars movies. For some questions, the respondent had to check one or more boxes. This type of data is difficult to represent in columnar format. As a result, this data set needs a lot of cleaning.

In this project we're going to clean and explore this data to see what we can find out.

## Overview our data 

```python
import pandas as pd
#We need to specify an encoding because the data set has some characters that aren't in Python's default utf-8 encoding
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
#view data
star_wars.head(10)
```

```python
# check star_wars columns
star_wars.columns
```

```python
pd.notnull(star_wars['RespondentID']).value_counts()
```

```python
# we keep only the rows where RespondentID isn't NaN 
star_wars = star_wars[pd.notnull(star_wars['RespondentID'])==True]
```

## Cleaning and Mapping Yes/No Columns

Above take a look at 2 columns :
- Have you seen any of the 6 films in the Star Wars franchise?
- Do you consider yourself to be a fan of the Star Wars film franchise?

Seeing the questions , it could be able to receive 2 good responses `Yes` or `No`.

```python
columns = ["Have you seen any of the 6 films in the Star Wars franchise?","Do you consider yourself to be a fan of the Star Wars film franchise?"]
for col in columns :
    print(star_wars[col].value_counts())
```

We're going to convert the columns from `String` type to `Boolean`. Why this choose, because using `Boolean` make the analysis easier.

```python
# map of our responses
responses = {"Yes":True,"No":False}
for col in columns :
    star_wars[col] = star_wars[col].map(responses)
# Columns check 
for col in columns :
    print(star_wars[col].value_counts())
```

## Cleaning and Mapping Checkbox Columns
The next six columns represent a single checkbox question. The question is `Which of the following Star Wars films have you seen? Please select all that apply.`
The columns for this question are:

- Which of the following Star Wars films have you seen? Please select all that apply. - Whether or not the respondent saw Star Wars: Episode I The Phantom Menace.
- Unnamed: 4 - Whether or not the respondent saw Star Wars: Episode II Attack of the Clones.
- Unnamed: 5 - Whether or not the respondent saw Star Wars: Episode III Revenge of the Sith.
- Unnamed: 6 - Whether or not the respondent saw Star Wars: Episode IV A New Hope.
- Unnamed: 7 - Whether or not the respondent saw Star Wars: Episode V The Empire Strikes Back.
- Unnamed: 8 - Whether or not the respondent saw Star Wars: Episode VI Return of the Jedi.

We are going to convert those columns to `Boolean` and renanme the columns to intuitive names.

```python
import numpy as np
responses = {
   "Star Wars: Episode I  The Phantom Menace": True,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope":True,
    "Star Wars: Episode V The Empire Strikes Back":True,
    "Star Wars: Episode VI Return of the Jedi":True, 
    np.NaN: False
}
# Convert to Boolean
for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(responses)

# Rename the columns
names = {
   "Which of the following Star Wars films have you seen? Please select all that apply.":"seen_1",
    "Unnamed: 4":"seen_2",
    "Unnamed: 5":"seen_3",
    "Unnamed: 6":"seen_4",
    "Unnamed: 7":"seen_5",
    "Unnamed: 8":"seen_6",
}
star_wars = star_wars.rename(columns=names)
star_wars.head(10)
```

## Cleaning the Ranking Columns
The next six columns ask the respondent to rank the Star Wars movies in order of least favorite to most favorite. Each of the following columns can contain the value `1`, `2`, `3`, `4`, `5`, `6`, or `NaN`. 
We are going to convert those columns to a `numeric` type and rename them to intuitive names.

```python
# Convert the columns to numeric
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
names = {
   "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.":"ranking_1",
    "Unnamed: 10":"ranking_2",
    "Unnamed: 11":"ranking_3",
    "Unnamed: 12":"ranking_4",
    "Unnamed: 13":"ranking_5",
    "Unnamed: 14":"ranking_6",
}

star_wars = star_wars.rename(columns=names)
# Check the result
star_wars.loc[:,"ranking_1":"ranking_6"].head(10)
```

## Finding the Highest-Ranked Movie
Now we've done cleaning the ranking columns , let's find the highest-ranked movie.

```python
means = star_wars.loc[:,"ranking_1":"ranking_6"].mean()
print(means)
# make a bar chat for our result
import matplotlib.pyplot as plt

plt.bar(range(6),means,color='r',tick_label=star_wars.columns[3:9])
```

By the result, the movie `seen_5` ( Star Wars: Episode V The Empire Strikes Back ) is the most popular of the Star Wars saga because it have the lower rank.

## Finding the Most Viewed Movie

```python
sums = star_wars.loc[:,"seen_1":"seen_6"].sum()
print(sums)

plt.bar(range(6),sums,color='r',tick_label=star_wars.columns[3:9])
```

The movie `seen_5` ( Star Wars: Episode V The Empire Strikes Back ) is the most viewed . We have this result because the most popular movie is obviously the most viewed, that's make sense. In fact, there is a strong correlation between watch a movie and rank it because it's after watching the movies you can appreciate them.

## Exploring the Data by Binary Segments

We are going to explore our data to have more information about this saga be relative to **Star Trek** . In our data , we have the column `Do you consider yourself to be a fan of the Star Trek franchise?` to know the fans of Star Trek

```python
reponses ={
    'Yes' : True,
    'No': False,
    np.NaN : False
}
# Convert the column to Boolean
star_wars["Do you consider yourself to be a fan of the Star Trek franchise?"] = star_wars["Do you consider yourself to be a fan of the Star Trek franchise?"].map(reponses)

#Split the data into two groups 
fans = star_wars[star_wars["Do you consider yourself to be a fan of the Star Trek franchise?"]== True]
not_Fans = star_wars[star_wars["Do you consider yourself to be a fan of the Star Trek franchise?"]== False]

labels = star_wars.columns[3:9]
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

# RANKS
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, fans.loc[:,"ranking_1":"ranking_6"].mean(), width, label='Fans')
rects2 = ax.bar(x + width/2, not_Fans.loc[:,"ranking_1":"ranking_6"].mean(), width, label='Not Fan')
# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Rank')
ax.set_title('Rank movies by fan of Star Trek or not')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()
plt.show()

# VIEW 
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, fans.loc[:,"seen_1":"seen_6"].mean(), width, label='Fans')
rects2 = ax.bar(x + width/2, not_Fans.loc[:,"seen_1":"seen_6"].mean(), width, label='Not Fan')
# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('View')
ax.set_title('View movies by fan of Star Trek or not')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()
plt.show()
#No_Star_Trek = star_wars[star_wars["Do you consider yourself to be a fan of the Star Trek franchise?"]== True]
```

We found something interresting, the people who gave most of the higher ranks to Star Wars are Star Trek's fans and who watch the most are them. This results show us how the fans of Star Trek love the saga of Star Wars. Finally we can say knowing Star Wars came out before Star Trek, Star Trek have success through the success of Star Wars because most of their fans love to watch Star Wars.

Let's check others columns , we can find another interresting results. 

### 1. Column : Education

We'll discover the education level of Star Wars's fans.

```python
# We'll save in this dictionary the frequency of "Education"
freq_edu ={}

for edu in star_wars["Education"]:
    # Test Nan value
    if pd.notnull(edu) == True:
        if edu in freq_edu :
            freq_edu[edu] += 1
        else:
            freq_edu[edu] = 1

freq_edu = pd.DataFrame([freq_edu])

freq_edu.plot(kind='bar',legend=True,figsize=(11,4))
```

Most of the fans of Star Wars are in `Bachelor degree` and `Some college or Associate degree` , the people who are not so interrested of it are `Less than high school degree`.

### 2. Column : Location (Census Region)

In this research we'll know the base area of Star Wars's fans

```python
# We'll save in this dictionary the frequency of "Location (Census Region) column"
freq_loc ={}

for loc in star_wars["Location (Census Region)"]:
    # Test Nan value
    if pd.notnull(loc) == True:
        if loc in freq_loc :
            freq_loc[loc] += 1
        else:
            freq_loc[loc] = 1

freq_loc = pd.DataFrame([freq_loc])

freq_loc.plot(kind='bar',legend=True,figsize=(11,4))
```

This plot show us the strong base area of the Star Wars fans , are `East North Central`, `Pacific` and `South Atlantic` . Where there a small base fans is `East South Central`.

Now we want to answer some question about Star Wars's characters:
>Which character do respondents like the most?

>Which character do respondents dislike the most?

>Which character is the most controversial (split between likes and dislikes)?

For that, we're going the work from the column 15 to 29 and see what we can find out .Let's take a look .

```python
star_wars.loc[:,star_wars.columns[15:29]].head(10)
```

```python
# rename the columns to intuitive names
names = {
   "Please state whether you view the following characters favorably, unfavorably, or are unfamiliar with him/her.":"character_1",
    "Unnamed: 16":"character_2",
    "Unnamed: 17":"character_3",
    "Unnamed: 18":"character_4",
    "Unnamed: 19":"character_5",
    "Unnamed: 20":"character_6",
    "Unnamed: 21":"character_7",
    "Unnamed: 22":"character_8",
    "Unnamed: 23":"character_9",
    "Unnamed: 24":"character_10",
    "Unnamed: 25":"character_11",
    "Unnamed: 26":"character_12",
    "Unnamed: 27":"character_13",
    "Unnamed: 28":"character_14",
}
star_wars = star_wars.rename(columns=names)

# check the differents answer in the columns
star_wars.loc[:,star_wars.columns[15:29]].value_counts()
```

We noticed as answer we have : `Very favorably`, `Neither favorably nor unfavorably (neutral)`, `Unfamiliar (N/A)`, `Somewhat favorably`. Now let's categorize them : 
 - Like : `Very favorably` and `Somewhat favorably`
 - Dislike : `Unfamiliar (N/A)` , 
 - Controversial : `Neither favorably nor unfavorably (neutral)`

```python
reponses ={
    'Very favorably' : 'Like',
    'Somewhat favorably' : 'Like',
    'Unfamiliar (N/A)' : 'Dislike',
    'Neither favorably nor unfavorably (neutral)' : 'Controversial',
    #np.NaN : "Controversial"
}
# Change the answers
for col in star_wars.columns[15:29] :
    star_wars[col] = star_wars[col].map(reponses)
    #print(len(star_wars[star_wars[col]=="Like"]))

star_wars.loc[:,star_wars.columns[15:29]]
#No_Star_Trek = star_wars[star_wars["Do you consider yourself to be a fan of the Star Trek franchise?"]== True]
```

```python
#Split the data into two groups 
like = star_wars[star_wars.loc[:,star_wars.columns[15:29]] == 'Like']
dislike = star_wars[star_wars.loc[:,star_wars.columns[15:29]] == 'Dislike']
controversial = star_wars[star_wars.loc[:,star_wars.columns[15:29]] == 'Controversial']

labels = star_wars.columns[15:29]
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

# RANKS
fig, ax = plt.subplots(figsize=(12,5))
rects1 = ax.bar(x - width/2, like.loc[:,star_wars.columns[15:29]].count(), width, label='Like')
rects2 = ax.bar(x + width/2, dislike.loc[:,star_wars.columns[15:29]].count(), width, label='Dislike')
rects3 = ax.bar(x + width, controversial.loc[:,star_wars.columns[15:29]].count(), width, label='Controversial')
# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Frequence')
ax.set_title('View about characters')
ax.set_xticks(x)
ax.set_xticklabels(labels,rotation=50)
ax.legend()

fig.tight_layout()
plt.show()
```

By this plot we can easily reply to our questions:
>Which character do respondents like the most?: character_2

>Which character do respondents dislike the most?: character_13

>Which character is the most controversial (split between likes and dislikes)?: character_9
